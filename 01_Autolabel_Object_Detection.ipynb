# relabel to corr model and type

```python
def load_xml(file):
    tree = ET.parse(file)
    return tree

def relabel_model_type(input_path, data_dict):

    for data in data_dict:
        for file in os.scandir(input_path):
            
            vin=str(pathlib.Path(file).name).split('_')[0][-7:]
            
            if file.name.endswith(".xml") and data["vin"] == vin:                
                xml_tree = load_xml(file)
                root = xml_tree.getroot()
                for obj in root.findall("object"):
                    for name in obj.findall("name"):
                        if name.text == "model": 
                            name.text = data["vehicle_type"].split(" ")[0]
                        if name.text == "type":
                            try:
                                name.text = data["vehicle_type"].split(" ")[1]
                            except:
                                print("no type", file)
                xml_tree.write(f"{input_path}/{os.path.basename(file)}")
```
